```python
import pandas as pd
data=pd.read_csv('data.csv', on_bad_lines='skip')
data.head()

```

```python
data['strength'].unique()
```

```python
data[data['password'].isnull()]
```

```python
password_tuple=np.array(data)
password_tuple
```

```python
import random
random.shuffle(password_tuple)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character

x=[labels[0] for labels in password_tuple]
y=[labels[1] for labels in password_tuple]


vectorizer=TfidfVectorizer(tokenizer=word_divide_char)  #Converts a collection of raw documents to a matrix of TF-IDF features.
X=vectorizer.fit_transform(x)
X.shape
vectorizer.get_feature_names_out()
```

```python
first_document_vector=X[0]
first_document_vector
first_document_vector.T.todense()
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names_out(),columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(random_state=0,multi_class='multinomial')
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
dt=np.array(['amar'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
dtc=DecisionTreeClassifier(random_state=0)
dtc.fit(X_train,y_train)
print(classification_report(y_test,y_pred))
```
